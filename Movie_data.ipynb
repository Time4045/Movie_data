```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
movie_data=pd.read_csv('/Users/maksimtrebusinin/Downloads/movie_metadata.csv')
```

```python
movie_data
```

```python
movie_data.columns
```

```python
movie_data['content_rating'].unique()
```

```python
re_movie_data=movie_data.drop(['color','director_name','director_facebook_likes','actor_2_name','gross',
                               'genres','actor_1_name','movie_title','actor_3_name','movie_imdb_link','language',
                              'country','plot_keywords','content_rating'], axis=1)
```

```python
re_movie_data=re_movie_data.fillna(0)
```

```python
re_movie_data
```

```python
re_movie_data['good_film']=np.where(re_movie_data.imdb_score > 7.0, 1, 0)
```

```python
re_movie_data
```

```python
X=re_movie_data.drop(['imdb_score','good_film'], axis=1)
y=re_movie_data['good_film']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.44, random_state=42)
```

```python
clf=RandomForestClassifier()
```

```python
parametrs={'n_estimators':[10,20,30,40], 'max_depth': range(1,10,2), 'min_samples_split':[2,4,6,8,9],
           'min_samples_leaf':[1,2,3,4]}
```

```python
grid_search_clf=GridSearchCV(clf, parametrs, cv=5)
```

```python
grid_search_clf.fit(X_train,y_train)
```

```python
best_clf_for_movie=grid_search_clf.best_estimator_
```

```python
grid_search_clf.best_params_
```

```python
re_movie_data['predictions']=best_clf_for_movie.predict(X)
```

```python
accuracy_score(y, re_movie_data['predictions'])
```

```python
movie_data['predictions']=re_movie_data['predictions']
```
